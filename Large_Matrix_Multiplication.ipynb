### Large Matrix Multiplication (Strassen's Algorithm)

### Randomly generating a large n * n square matrix

```python
import numpy as np
import random 
A = np.random.random((16, 16))
B = np.random.random((16, 16))
```

### Strassen's Algorithm


```python
import time
start = time.perf_counter()
if A.shape[0] == len(B[0]):
    #Strassen's algorithm
    n = len(A)
    def adding (A, B):
        for i in range(n):
            for j in range(n):
                add_matrix = A[i][j] + B[i][j]
        return add_matrix
    
    def subtracting (A, B):
        for i in range(n):
            for j in range(n):
                lesser_matrix = A[i][j] - B[i][j]
        return lesser_matrix
    
    #Recursive Step
    def split(X):
        n0 = len(X)
        s = n0 // 2
        A11 = [row[:s] for row in X[:s]]
        A12 = [row[s:] for row in X[:s]]
        A21 = [row[:s] for row in X[s:]]
        A22 = [row[s:] for row in X[s:]]
        return A11, A12, A21, A22
    
    def glue(C11, C12, C21, C22):
        for c11, c12 in zip(C11, C12):
            top = c11 + c12
        for c21, c22 in zip(C21, C22):
            bott = c21 + c22
        return top + bott


    #Conquer Step
    def strassen (A, B):
        n = len(A)
    
        if n == 1:
            return [[A[0][0] * B[0][0]]]
        
        #Splitting Matrices
        A11, A12, A21, A22 = split(A)
        B11, B12, B21, B22 = split(B)

        #strassen it
        p1 = strassen(A11, subtracting(B12, B22))
        p2 = strassen(adding(A11, A12), B22)
        p3 = strassen(adding(A21, A22), B11)
        p4 = strassen(A22, subtracting(B21, B11))
        p5 = strassen(adding(A11, A12),adding(B11, B12))
        p6 = strassen(subtracting(A12, A22), adding(B21, B22))
        p7 = strassen(subtracting(A11, A21), adding(B11, B12))

        # Multiplied matrix
        C11 = adding(subtracting(adding(p5, p4), p2), p6)
        C12 = adding(p1, p2)
        C21 = adding(p3, p4)
        C22 = subtracting(subtracting(adding(p1, p5), p3), p7)

        #Glueing it together
        new_matrix = glue(C11, C12, C21, C22)

        return new_matrix
        
   
else:
    print("Not compatible for multiplication")

end = time.perf_counter()

total_time = end - start
print(f'Total implementation time = {total_time: 3f} seconds')
    
    


```

### Classical Matrix Multiplication


```python
import time
start = time.perf_counter()

def matrix_multiplication(A, B):
    rows_A, cols_A = len(A), len(A[0])
    rows_B, cols_B = len(B), len(B[0])

    #check if multiplication is possible
    if rows_A == cols_B:
        result = [[0 for _ in range(cols_B)] for _ in range(rows_A)]
        for i in range(rows_A):
            for j in range(cols_B):
                for k in range(cols_A):
                    result [i][j] += A[i][k] * B[k][j]
        return result
    else:
        print("Multiplication is invalid")

print(matrix_multiplication(A, B))

end = time.perf_counter()

total = end - start
print("Implementation time = ", total)
```
